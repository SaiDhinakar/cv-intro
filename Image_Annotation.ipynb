```python
image = cv.imread('/content/person.jpg')
plt.imshow(cv.cvtColor(image, cv.COLOR_BGR2RGB))
image.shape
```

Draw line

```python
line = cv.line(image.copy(), (0,0), (image.shape[1], image.shape[0]), (0, 0, 255), 3)
plt.imshow(cv.cvtColor(line, cv.COLOR_BGR2RGB))
```

Draw a Rectangle

```python
rect = cv.rectangle(image.copy(), (250, 50), (450, 300), (0, 255, 0), 2)
plt.imshow(cv.cvtColor(rect, cv.COLOR_BGR2RGB))
```

Text

```python
res_image = cv.putText(rect.copy(), "Peter Parker", (50, 100), cv.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 225), 2)
plt.imshow(cv.cvtColor(res_image, cv.COLOR_BGR2RGB))
```
